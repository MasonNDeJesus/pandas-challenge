```python
# Imports and what not
import pandas as pd
from pathlib import Path

# Files
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# I just kept this stuff here cuz I was unsure of what purpose it served in the homework as a whole.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# District Summary

```python
# Total number of schools using .unique to determine the amount by finding unique names within the set.
total_unique_schools = school_data['school_name'].nunique()
print(total_unique_schools)
```

```python
# Total students using a simple .count function
total_students = student_data['Student ID'].count()
print(total_students)
```

```python
# Total budget using a simple sum function
total_budget = school_data['budget'].sum()
print(total_budget)
```

```python
# Avegare math score calculated by using the .mean function
average_math_score = student_data['math_score'].mean()
print(average_math_score)
```

```python
# Avegare reading score calculated by using the .mean function
average_reading_score = student_data['reading_score'].mean()
print(average_reading_score)
```

```python
# Passing math count using the .sum function and finding scores that are above or equal to 70
# Percent passing math used by simply dividing the passing math count by the total students, then multiplying by 100
passing_math_count = (student_data['math_score'] >= 70).sum()
percent_passing_math = (passing_math_count / total_students) * 100
print(percent_passing_math)
```

```python
# Passing reading count using the .sum function and finding scores that are above or equal to 70
# Percent passing math used by simply dividing the passing reading count by the total students, then multiplying by 100
passing_reading_count = (student_data['reading_score'] >= 70).sum()
percent_passing_reading = (passing_reading_count / total_students) * 100
print(percent_passing_reading)
```

```python
# The amount people passing both is determined by simply using & and .sum to combine the passing math and passing reading count
# The percent that are passing both is then found by dividing the passing both count by the total students, then multiplying by 100
passing_both_count = ((student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)).sum()
percent_passing_both = (passing_both_count / total_students) * 100
print(percent_passing_both)
```

```python
# Formatting using the stuff found in the example
total_students_formatted = '{:,}'.format(total_students)
total_budget_formatted = '${:,.2f}'.format(total_budget)

# Dataframe is used to encompass all the data into a single variable, that being district_summary, then I used .head() to finish it off
district_summary = pd.DataFrame({
    'Total Unique Schools': [total_unique_schools],
    'Total Students': [total_students_formatted],
    'Total Budget': [total_budget_formatted],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [percent_passing_math],
    '% Passing Reading': [percent_passing_reading],
    '% Overall Passing': [percent_passing_both]
})
district_summary.head()
```

# School Summary

```python
# Used the function from the example. 
# From this point foward I will not be printing the info anymore as there doesn't seem to be a requirement to do so.
# If it is needed after grading, I just have to simply add print functions to everything
school_types = school_data.set_index(["school_name"])["type"]
```

```python
# Used the .groupby and .nunique functions to find the amount of students attending each school
per_school_counts = school_data_complete['school_name'].value_counts()
```

```python
# Total school budget and per capita spending per school. Assuming the per capita spending is based on the size, this is the following code I made.
per_school_budget = school_data.groupby("school_name")["budget"].sum()
per_school_capita = per_school_budget/per_school_counts
```

```python
# Using a new function called .reset_index(), which resets the index to allow for the data to be merged with other data that doesn't have the same structure
per_school_math = school_data_complete.groupby('school_name')['math_score'].mean()
per_school_reading = school_data_complete.groupby('school_name')['reading_score'].mean()
```

```python
# A bit more complex as a whole. The second part of this code was a lot longer due to the fact that I needed to find something specific within the data
# I'm sure there was an easier way, but this still provides the results needed.
students_passing_math = (school_data_complete['math_score'] >= 70).sum()
school_students_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()
```

```python
# I just copied the above code and made some changes to it.
students_passing_reading = (school_data_complete['reading_score'] >= 70).sum()
school_students_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()
```

```python
# Used the code as was told in the example
students_passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
```

```python
# This stuff kicked the ever-living crap out of me. I played around a ton with the code and eventually, after some changes, got it to work
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})

# Formatting
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map('${:,.2f}'.format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map('${:,.2f}'.format)

# Ran into a problem with displaying all of the schools, so set the max displayed rows to None, which allows all rows to be shown
# Using with keeps the lack of a limit on max rows only to this line of code, that way I don't have to walk it back later
with pd.option_context('display.max_rows', None):
    
    display(per_school_summary)
```

```python
# Used the provided code for establishing the bins
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Remove formatting from copy. It was causing a lot of problems for the pd.cut function so I played around a bit and came up with this
numeric_columns = ["Per Student Budget"]
for col in numeric_columns:
    school_spending_df[col] = school_spending_df[col].str.replace("$","").str.replace(",","")
school_spending_df[numeric_columns] = school_spending_df[numeric_columns].apply(pd.to_numeric)
```

```python
# I hated this. Absolutely hated this. pd.cut was awful to use and I ran into error after error with this crap.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins = spending_bins, labels = spending_labels)

# Bringing back the formatting
school_spending_df['Per Student Budget'] = school_spending_df['Per Student Budget'].map('${:,.2f}'.format)

# Displaying the dataframe
school_spending_df.head(total_unique_schools)
```

```python
# The provided code did not work this time. I had to do a bit of research and found a way to get around the error preventing me from moving forward
# Observed=False was what I used to get around the error, as it seemed to be an error with .groupby. I suppressed the warning and got the right results.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=False)["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=False)["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=False)["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=False)["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=False)["% Overall Passing"].mean()
```
